```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips= sns.load_dataset('tips')
tips
```

1. heat map

```python
tipscorr = tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

2. cluster map

```python
sns.clustermap(tipscorr.corr())
```

pivot table heat map

```python
pvtflight = flights.pivot_table(values='passengers',index='year',columns='month')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
