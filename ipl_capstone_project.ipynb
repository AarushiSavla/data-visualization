# some basic questions 

1. which team won the most matches 

```python
df.head()
```

```python
match_wins=df['match_winner'].value_counts()
match_wins
```

```python
sns.barplot(y =match_wins.index, x=match_wins.values, palette='viridis' )
plt.title("most match win by team")
```

2. toss desicion trends 

```python
df.head()
```

```python
# create a countplot of toss desicions 
sns.countplot(x=df['toss_decision'],palette='magma')
plt.title("toss desicion trends ")
```

3. toss winner vs match winner 

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage = (count *100)/df.shape[0]
percentage
```

4. how do teams win ? (runs vs wickets)

```python
fig = px.histogram(df, x='won_by', title='Match Wins by Category')
fig.show()
```

```python
sns.countplot(x= df['won_by'],palette='plasma')
plt.title('Match Wins by Category')
```

```python
high.plot(kind='barh')
```

10 best bowling figures 


```python
df.head()
```

```python
# extract only the wickets from the best_bowling_figure
df['highest_wicket']=df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wicket']=df['highest_wicket'].astype(int)
df.head()
```

```python
best_bowling_fig = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
best_bowling_fig
```

```python
best_bowling_fig.plot(kind='barh')
```
