```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import cufflinks as cf 
cf.go_offline()
cf.set_config_file(theme='white', offline=True)
from plotly.offline import iplot
```

```python
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# creating a linegraph 
fig = px.line(tips, y='total_bill', title='Total Bill Amount')
fig.show()
```

```python
# barplt
fig=px.bar(tips.groupby('day')['tip'].mean())
fig.show()
```

```python
fig=px.scatter(tips,x='total_bill',y='tip')
fig.show()
```

```python
fig=px.box(tips,x='total_bill',y='day')
fig.show()
```
