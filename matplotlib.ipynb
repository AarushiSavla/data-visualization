# matplot lib basics 

```python
# numpy array - linspace - linear space 
x = np.linspace(0,5,11)
x
```

```python
y = x ** 3
y
```

```python
#plot - makes a line graph 
plt.plot(x,y)
# graph title
plt.title("hello ji")
# x label 
plt.xlabel("blah blue")
#y label
plt.ylabel("bleh blue")
```
