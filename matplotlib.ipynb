# data visualization 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
```

# matplot lib basics 

```python
# numpy array - linspace - linear space 
x = np.linspace(0,5,11)
x
```

```python
y = x ** 3
y
```

```python
#plot - makes a line graph 
plt.plot(x,y)
# graph title
plt.title("hello ji")
# x label 
plt.xlabel("blah blue")
#y label
plt.ylabel("bleh blue")
```

# subplots in matplotlib 

```python
# ek sath multiple plots 
# you need to provide rows and columns 
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
```

```python
plt.subplot(6,6,1)
plt.subplot(6,6,2)
plt.subplot(6,6,3)
plt.subplot(6,6,4)
plt.subplot(6,6,5)
plt.subplot(6,6,6)
plt.subplot(6,6,7)
plt.subplot(6,6,30)
plt.subplot(6,6,36)
```

# OBJECT ORIENTED WAY 

```python
# CREATE AN OBJECT 
fig = plt.figure(figsize=(9,8),dpi=100)
# ADD AXES -  PROVIDE 4 THINGS 
#  axes_name = fig.add_axes([left, bottom, width, height])
axis1 = fig.add_axes([0.1,0.1,1,1])
#plotting on axis1 
axis1.plot(x,y)
#create axis2 
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

 SAVING PLOTS 

```python
#FIGURE
fig = plt.figure()
#ADD AXES 
ax=fig.add_axes([0.1,0.1,0.8,0.8])
# PLOT 
ax.plot(x,y)
#SAVE 
plt.savefig("basicplot.png")
```
