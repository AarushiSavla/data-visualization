```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# loading csv file from seaborn github
df = sns.load_dataset('tips')
df
```

```python
# size column will also be consided categorical or quantitative
df['size'].unique()
```

SOME TYPES OF DISTRIBUTION PLOTS 
- distplot/histplot
- jointplot
- pairplot
- rugplot

1. distplot 

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

2. jointplot

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df ,kind='reg')
```

3. pairplot

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

4. rugplot

```python
sns.rugplot(df)
```
